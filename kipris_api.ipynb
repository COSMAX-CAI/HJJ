```python
import requests
import xmltodict
import pandas as pd
```

```python
def get_url(word, years, rows, patent, utility):
    base = "http://plus.kipris.or.kr/kipo-api/kipi/patUtiModInfoSearchSevice/"
    param_word = "getWordSearch?word=" + str(word)
    param_years = "&year=" + str(years)
    param_rows = "&numOfRows=" + str(rows)
    param_patent = "&paten=" + str(patent)
    param_utility = "&utility=" + str(utility)
    key = "&ServiceKey=kGFNhQGiCDQLoNH9=tROJeME7zE0aP419nT9zFhRcqQ="
    
    return base + param_word + param_years + param_rows + param_patent + param_utility + key
```

```python
get_url("화장료", "5", "100", "True", "True")
```

```python
response = requests.get(get_url("화장료", "5", "100", "True", "True"))
content = response.content
dict_type = xmltodict.parse(content)
body = dict_type['response']['body']
items = body['items']['item']
```

```python
df = pd.DataFrame(columns = range(16))
df.columns = items[0].keys()
df = df.set_index('indexNo', drop=True)
```

```python
def get_data(x):
    data_list = []
    for i in range(len(items)):
        data = items[i].get(x)
        data_list.append(data)
    return data_list

def add_data(x):
    df[x] = get_data(x)
```

```python
for i in df.columns:
    add_data(i)
```
